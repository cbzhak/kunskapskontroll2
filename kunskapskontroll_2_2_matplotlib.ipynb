### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

De två sätten är:
1. Pyplot-gränssnittet (matplotlib.pyplot). Detta liknar MATLAB, enkelt för snabba plottar.
2. Objektorienterat gränssnitt (OO), där man arbetar direkt med Figure och Axes-objekt.

Rekommenderat sätt:
Det objektorienterade gränssnittet rekommenderas, särskilt vid mer avancerade eller anpassade grafer.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: hela figurfönstret eller sidan som innehåller allt.


Axes: ett enskilt diagram eller subplot inom figuren.


Axis: enskild x- eller y-axel i ett Axes-objekt.


Artist: allt som ritas upp och syns, t.ex. linjer, text, etiketter och axlar.

### When plotting in Matplotlib, what is the expected input data type?

Matplotlib förväntar sig array-liknande data, t.ex. Python-listor, NumPy-arrayer, eller Pandas-serier.
Intern omvandling sker till NumPy-arrayer vid plotting.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt

# Skapa x-värden från -4 till 4, med 100 jämnt fördelade punkter
x = np.linspace(-4, 4, 100)
y = x**2  # y = x²

# Pyplot-sättet
plt.plot(x, y)
plt.title("y = x²")          # Titel
plt.xlabel("x")              # X-axel etikett
plt.ylabel("y")              # Y-axel etikett
plt.grid(True)               # Rutnät
plt.show()                   # Visa grafen

# Objektorienterat sätt
fig, ax = plt.subplots()     # Skapa figur och ax-objekt
ax.plot(x, y)
ax.set_title("y = x²")       # Titel
ax.set_xlabel("x")           # X-axel etikett
ax.set_ylabel("y")           # Y-axel etikett
ax.grid(True)                # Rutnät
plt.show()                   # Visa grafen
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Skapa figur med två subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  # 1 rad, 2 kolumner

# Subplot 1: scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.set_title("Scatter plot")
ax1.set_xlabel("X-värden")
ax1.set_ylabel("Y-värden")

# Subplot 2: stapeldiagram
ax2.bar(names, values, color='green')
ax2.set_title("Fruktmängd")
ax2.set_ylabel("Antal")

plt.tight_layout()  # Justera layouten så inget överlappar
plt.show()
```
